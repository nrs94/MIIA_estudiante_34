## Pregunta 1

Con la base de datos *COVID-19.csv* crea un dashboard en `panel` que:

1. Permita crear un diagrama de barras horizontal que muestre los casos de COVID-19 de acuerdo a la selección de los siguentes criterios:<br />
    a. Casos o muertes.<br />
    b. Periodo: Histórico, Últimos 7 días y Últimas 24 horas.<br />
    Pista: utilice un control de tipo `RadioButtonGroup` para cada uno de los dos criterios.<br />
2. Contenga un diagrama de barras horizontal de las muertes por millón de personas.
3. Permita visualizar a tráves de un diagrama de barras horizontal los casos en cada una de las regiones de la Organización Munidal de la Salud.

Nota: a excpeción de las regiones, muestre los 10 mayores valores en cada gráfico.

| <center>Variable</center> |<center>Descripción</center> | 
| :-- | :-- | 
| Name | País |
| WHO Region | Región de la OMS |
| Cases - cumulative total | Total de casos hasta el momento |
| Cases - cumulative total per 1 million population | Número de casos acumulado por millón de personas |
| Cases - newly reported in last 7 days | Número de casos nuevos en los últimos 7 días |
| Cases - newly reported in last 24 hours | Número de casos nuevos en las últimas 24 horas |
| Deaths - cumulative total | Total de muertes hasta el momento |
| Deaths - cumulative total per 1 million population | Número de muertes acumulado por millón de personas |
| Deaths - newly reported in last 7 days | Número de muertes nuevas en los últimos 7 días |
| Deaths - newly reported in last 24 hours | Número de muertes nuevas en las últimas 24 horas |

```python
# YOUR CODE HERE
df = pd.read_csv('Archivos/COVID-19.csv')
df.head()
```

```python
# YOUR CODE HERE
titulo = '# Covid 19'
```

```python
# YOUR CODE HERE
seleccion1 = pn.widgets.RadioButtonGroup(name='Tipo', options= ['Casos','Muertes'])
seleccion2 = pn.widgets.RadioButtonGroup(name='Periodo', options= ['Histórico','Últimos 7 días','Últimas 24 horas'])
```

```python
# YOUR CODE HERE
# YOUR CODE HERE

@pn.depends(seleccion1,seleccion2)


def funcion_interactiva(seleccion1,seleccion2):

    if (seleccion1 == "Casos") and (seleccion2 == 'Histórico'):
        df_filtered = df[['Name','Cases - cumulative total']]
        df_filtered = df_filtered.sort_values(by=['Cases - cumulative total'],ascending=False)
        df_filtered = df_filtered.head(10)
        df_filtered = df_filtered.sort_values(by=['Cases - cumulative total'],ascending=True)
        df_filtered = df_filtered.rename(columns={'Cases - cumulative total':'Datos'})
    elif (seleccion1 == "Casos") and (seleccion2 == 'Últimos 7 días'):
        df_filtered = df[['Name','Cases - newly reported in last 7 days']]
        df_filtered = df_filtered.sort_values(by=['Cases - newly reported in last 7 days'],ascending=False)
        df_filtered = df_filtered.head(10)
        df_filtered = df_filtered.sort_values(by=['Cases - newly reported in last 7 days'],ascending=True)
        df_filtered = df_filtered.rename(columns={'Cases - newly reported in last 7 days':'Datos'})
    elif (seleccion1 == "Casos") and (seleccion2 == "Últimas 24 horas"):
        df_filtered = df[['Name','Cases - newly reported in last 24 hours']]
        df_filtered = df_filtered.sort_values(by=['Cases - newly reported in last 24 hours'],ascending=False)
        df_filtered = df_filtered.head(10)
        df_filtered = df_filtered.sort_values(by=['Cases - newly reported in last 24 hours'],ascending=True)
        df_filtered = df_filtered.rename(columns={'Cases - newly reported in last 24 hours':'Datos'})
    elif (seleccion1 == "Muertes") and (seleccion2 == 'Histórico'):
        df_filtered = df[['Name','Deaths - cumulative total']]
        df_filtered = df_filtered.sort_values(by=['Deaths - cumulative total'],ascending=False)
        df_filtered = df_filtered.head(10)
        df_filtered = df_filtered.sort_values(by=['Deaths - cumulative total'],ascending=True)
        df_filtered = df_filtered.rename(columns={'Deaths - cumulative total':'Datos'})
    elif (seleccion1 == "Muertes") and (seleccion2 == 'Últimos 7 días'):
        df_filtered = df[['Name','Deaths - newly reported in last 7 days']]
        df_filtered = df_filtered.sort_values(by=['Deaths - newly reported in last 7 days'],ascending=False)
        df_filtered = df_filtered.head(10)
        df_filtered = df_filtered.sort_values(by=['Deaths - newly reported in last 7 days'],ascending=True)
        df_filtered = df_filtered.rename(columns={'Deaths - newly reported in last 7 days':'Datos'})
    elif (seleccion1 == "Muertes") and (seleccion2 == 'Últimas 24 horas'):
        df_filtered = df[['Name','Deaths - newly reported in last 24 hours']]
        df_filtered = df_filtered.sort_values(by=['Deaths - newly reported in last 24 hours'],ascending=False)
        df_filtered = df_filtered.head(10)
        df_filtered = df_filtered.sort_values(by=['Deaths - newly reported in last 24 hours'],ascending=True)
        df_filtered = df_filtered.rename(columns={'Deaths - newly reported in last 24 hours':'Datos'})
        
    #df_filtered = df_filtered.head(10)
    
    
    # Definimos y retornamos la figura.
    fig = go.Figure(go.Bar(
                            x=df_filtered['Datos'],
                            y=df_filtered['Name'],
                            textposition='outside',                                
                            marker_color='rgb(26, 118, 255)',                      
                            orientation='h'                                        
                          )
                   )
    
                    
    fig.update_layout(
                        barmode='stack',                                           
                        yaxis_title="Name",
                        xaxis_title="Conteo",
                        title= 'Distribución Casos y Muertes',
                        xaxis=dict(showgrid=False),
                        yaxis=dict(showgrid=False),                               
                        plot_bgcolor='white',                                      
                        showlegend=False                                          
                     )
    
    return fig

def muertes_millon():
    df_filtered = df[['Name','Deaths - cumulative total per 1 million population']]
    df_filtered = df_filtered.sort_values(by=['Deaths - cumulative total per 1 million population'],ascending=False)
    df_filtered = df_filtered.head(10)
    df_filtered = df_filtered.sort_values(by=['Deaths - cumulative total per 1 million population'],ascending=True)
    
    fig = go.Figure(go.Bar(
                            x=df_filtered['Deaths - cumulative total per 1 million population'],
                            y=df_filtered['Name'],
                            textposition='outside',                                
                            marker_color='rgb(26, 118, 255)',                      
                            orientation='h'                                        
                          )
                   )
    
                    
    fig.update_layout(
                        barmode='stack',                                           
                        yaxis_title="Name",
                        xaxis_title="Conteo",
                        title= 'Muertes por millón de personas',
                        xaxis=dict(showgrid=False),
                        yaxis=dict(showgrid=False),                               
                        plot_bgcolor='white',                                      
                        showlegend=False                                          
                     )
    
    return fig

def casos_region():
    
    df_filtered = pd.DataFrame(df.groupby('WHO Region')['Cases - cumulative total'].sum())
    df_filtered = df_filtered.sort_values(by=['Cases - cumulative total'])
    
    fig = go.Figure(go.Bar(
                            x=df_filtered['Cases - cumulative total'],
                            y=df_filtered.index,
                            textposition='outside',                                
                            marker_color='rgb(26, 118, 255)',                      
                            orientation='h'                                        
                          )
                   )
    
                    
    fig.update_layout(
                        barmode='stack',                                           
                        yaxis_title="Name",
                        xaxis_title="Conteo",
                        title= 'Casos por región',
                        xaxis=dict(showgrid=False),
                        yaxis=dict(showgrid=False),                               
                        plot_bgcolor='white',                                      
                        showlegend=False                                          
                     )
    
    return fig
```

```python
tablero_control = pn.Row(pn.Column(titulo, seleccion1, seleccion2, funcion_interactiva), pn.Column(muertes_millon,casos_region))
```

```python
# YOUR CODE HERE
tablero_control.servable()
```

```python
# YOUR CODE HERE
tablero_control.show()
```
