```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tools.eval_measures import rmse
import plotly.graph_objects as go
import ipywidgets as widgets
from ipywidgets import interactive
```

```python
df1 = pd.read_csv('Archivos/SB11_20001.txt',sep="|",encoding='latin-1',dtype='unicode')
df2 = pd.read_csv('Archivos/SB11_20011.txt',sep="|",encoding='latin-1',dtype='unicode')
df3 = pd.read_csv('Archivos/SB11_20021.txt',sep="|",encoding='latin-1',dtype='unicode')

df = df1.append([df2,df3])

list_puntos = ['PUNT_BIOLOGIA',
               'PUNT_MATEMATICAS',
               'PUNT_FILOSOFIA',
               'PUNT_FISICA',
               'PUNT_HISTORIA',
               'PUNT_QUIMICA',
               'PUNT_LENGUAJE',
               'PUNT_GEOGRAFIA',
               'PUNT_INTERDISCIPLINAR',
               'PUNT_IDIOMA']

for subject in list_puntos:
    df[subject] = pd.to_numeric(df[subject], errors='coerce',downcast='integer')
    
df['PUNTOS_FINAL'] = df[list_puntos].sum(axis=1)
```

### Información de los datos

Para este ejercicio contamos con 3 set de datos de las Pruebas Saber 11 de los años y semestres 2000-1,2001-1, 2002-1 con un total de 218958 registros. Para cada uno de este set de datos contamos con 57 diferentes variables de diferentes tipos. 

Tenemos variables cuantitativas como el puntaje de los estudiantes en diferentes asignaturas, fecha de nacimiento etc. Adicionalmente, tenemos variables cualitativas de tipo nominal como el sexo del estudiante, jornada de colegio, departamento, etc, y de tipo ordinal como la educación de padre, de la madre entre otras.

```python
df_subset = df[['PUNTOS_FINAL','COLE_JORNADA','ESTU_GENERO','COLE_VALOR_PENSION','FAMI_NUM_HERMANOS','ESTU_TRABAJA','COLE_CALENDARIO','FAMI_INGRESO_FMILIAR_MENSUAL']]

df_subset['COLE_VALOR_PENSION'] = df_subset['COLE_VALOR_PENSION'].replace(['*','@'],None)
df_subset['FAMI_NUM_HERMANOS'] = df_subset['FAMI_NUM_HERMANOS'].replace(['*','@'],None)
df_subset['FAMI_NUM_HERMANOS'] = df_subset['FAMI_NUM_HERMANOS'].replace(['+'],10)
df_subset['ESTU_TRABAJA'] = df_subset['ESTU_TRABAJA'].replace(['*','@'],None)
df_subset['FAMI_INGRESO_FMILIAR_MENSUAL'] = df_subset['FAMI_INGRESO_FMILIAR_MENSUAL'].replace(['*','@'],None)

df_subset['COLE_JORNADA'] = df_subset['COLE_JORNADA'].replace(['COMPLETA U ORDINARIA', 'MAÑANA', 'NOCHE', 'SABATINA-DOMINICAL', 'TARDE'],['C', 'M', 'N', 'S', 'T'])

df_subset['COLE_VALOR_PENSION'] = pd.to_numeric(df_subset['COLE_VALOR_PENSION'],errors='coerce',downcast='integer')
df_subset['FAMI_NUM_HERMANOS'] = pd.to_numeric(df_subset['FAMI_NUM_HERMANOS'],errors='coerce',downcast='integer')
df_subset['FAMI_INGRESO_FMILIAR_MENSUAL'] = pd.to_numeric(df_subset['FAMI_INGRESO_FMILIAR_MENSUAL'],errors='coerce',downcast='integer')
```

El entregable de esta misión es un reporte sobre el desempeño de los modelos propuestos para abordar al menos una de las preguntas guía planteadas, acompañado de una conclusión sobre los resultados del modelo (si son válidos) en el contexto de la problemática planteada.

```python
# Código
X = df_subset.drop("PUNTOS_FINAL", axis=1)
X = pd.get_dummies(X,drop_first=True)
Y = df_subset['PUNTOS_FINAL']

X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()
model.summary()
```

```python
#CALCULAR MEAN SQUARED ERROR
y_pred = model.predict(X)
mse = (rmse(Y,y_pred))**2
print(mse)
```

Para esto es necesario transformar las variables categóricas en variables dummies y dejando alguna categoria como base para la interpretación de resultados.

## Rendimiento del modelo

- Significancia Global: vemos en la salida anterior que el p-value de la prueba F es menor a 5%. Por lo tanto, con un nivel de significancia del 95% el modelo es globalmente significativo. Esto quiere decir, que al menos alguna variable si explica el puntaje obtenido por los estudiantes.

- Significancia Individual: observamos que todas las variables individualmente son significativas ya que su p-value es menor a 5% excepto COLE_CALENDARIO_B. Esto significa, que con un nivel de significancia del 95% estas variables son significativas.

- R2 ajustado: el 27.3% de la variación en la variable PUNTOS_FINAL se explica por la variación de las variables explicatorias. Este R2 ajustado puede ser bajo en términos generales y nos puede indicar que nuestro modelo no se ajusta bien a los datos. Sin embargo, no existe una regla general para el R2 ajustado y va a depender del área de estudio y de otros factores

```python
# Código
variable_widget = widgets.Dropdown(
                        description = 'Select Variable:',
                        options = ["COLE_JORNADA","ESTU_GENERO","COLE_VALOR_PENSION","FAMI_NUM_HERMANOS","ESTU_TRABAJA","COLE_CALENDARIO","FAMI_INGRESO_FMILIAR_MENSUAL"],
                        style = {'description_width': 'initial'}
                        )
filtro_widget = widgets.Dropdown(
                        description = 'Select Filter:',
                        style = {'description_width': 'initial'}
                        )

def on_update_filtro_widget(*args):
    valor = variable_widget.value
    filtro_widget.options = list(set(df_subset[valor]))
    
variable_widget.observe(on_update_filtro_widget, 'value')
```

```python
fig = go.FigureWidget(
                        data = [go.Histogram(name='Histograma')],  
    
                        layout = go.Layout(
                                            title = 'Distribución',
                                            xaxis_title = '',
                                            yaxis_title = 'Conteo',
                                          )
                     )
```

```python
def filtrar():
    fig.data[0].x = df_subset["PUNTOS_FINAL"][df_subset[variable_widget.value] == filtro_widget.value]
```

```python
def respuesta(change):
    filtrar()
```

```python
filtro_widget.observe(respuesta, 'value')
```

```python
widgets_filtros = widgets.HBox([variable_widget,filtro_widget])
```

```python
visualizacion = widgets.VBox(
              [widgets_filtros,
              fig]
            )
```

```python
display(visualizacion)
```
