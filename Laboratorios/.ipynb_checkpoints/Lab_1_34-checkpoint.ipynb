```python
# código
#lo que hago acá es leer cada archivo y crear un diccionario dónde en cada key guardo una lista con la info de cada archivo
def archivo_a_lista(archivo):
    directorio = "./Archivos/"+archivo+".txt"
    with open(directorio, "r") as f:
        f = f.read()
        lista = f.splitlines()
    return lista

diccionario = {}

archivos_string = ['escolaridad','estado_civil','estrato','genero','region']
for archivo in archivos_string:
    diccionario[archivo] = archivo_a_lista(archivo)

#ya que los archivos son txt venían como string entonces ciertas variables las transformo en númericas
archivos_number = ['edad','promedio']
for archivo in archivos_number:
    diccionario[archivo] = [float(x) for x in archivo_a_lista(archivo)]
    
diccionario['escolaridad']
```

```python
def mediana(lista):
    try:
        lista_mediana = lista.copy()
        lista_mediana.sort()
        mid = len(lista_mediana) // 2
        res = (lista_mediana[mid] + lista_mediana[~mid]) / 2
        return res
    except:
        return 'No es posible calcular mediana por que es una variable cualitativa'

# función que calcula la desviación de una lista de datos numéricos
def desviacion_estandar(lista):
    try:
        n = len(lista)
        mean = sum(lista)/n
        variance = sum([((x - mean) ** 2) for x in lista]) / n
        std = variance ** 0.5
        return round(std,2)
    except: 
        return 'No es posible calcular desviación estandar por que es una variable cualitativa'
    

def moda(lista):
    return max(set(lista), key=lista.count)

def promedio(lista):
    try:
        return round(sum(lista)/len(lista),2)
    except:
        return 'No es posible calcular desviación estandar por que es una variable cualitativa'

def conteos(lista):
    return dict((x,lista.count(x)) for x in set(lista))
```

```python
#esta funcion permite ver la distribución de una variable en especifico y diferentes métricas
#el primer input es el nombre
#el segundo input es el tipo de variable

def descripcion_variable(variable,tipo_variable):
    
    '''
    El primer input es el nombre de la variable entre comillas
    El segundo input es el tipo de variable puede tomar el valor de 'numerica' o 'categorica'
    '''
    
    variable = diccionario[variable]
    variable.sort()
    
    #gráfico
    plt.hist(variable)
    plt.title('Distribución variable')
    plt.ylabel('Conteo')
    plt.xlabel('Valores variable')
    plt.show()
    
    if tipo_variable == 'numerica':
        
        plt.boxplot(variable,vert=False)
        plt.show()
        
        print('El promedio de la variable es: {}'.format(promedio(variable)))
        print('La desviación estandar de la variable es: {}'.format(desviacion_estandar(variable)))
        print('La mediana de la variable es: {}'.format(mediana(variable)))
        print('La moda de la variable es: {}'.format(moda(variable)))
        print('El conteo de la variable es: {}'.format(conteos(variable)))
    elif tipo_variable == 'categorica':
        print('La moda de la variable es: {}'.format(moda(variable)))
        print('El conteo de la variable es: {}'.format(conteos(variable)))
    return 

#El input de la funcion descripción variable debe ser algún key de la variable diccionario
print(diccionario.keys())
descripcion_variable('edad','numerica')
```

**Misión 4**

Desarrollar un algoritmo que permita conocer la población que sería becada según los criterios propuestos por el *primer grupo* mencionado arriba.

```python
#creo unas listas por estrato con los identificadores de cada persona ordenados para saber cual escoger primero.
def get_ordered_ids(estrato):
    
    '''Función que genera una lista con los ids ordenados de mayor a menor promedio.
    El input es el estrato a evaluar.'''
    
    new_dic = {}
    for key, value in diccionario_new.items():
       if value['estrato'] == estrato:
           new_dic[key] = value
         
    lista_promedios = []
    for key, value in new_dic.items():
           lista_promedios.append(value['promedio'])
           
    lista_promedios = list(set(lista_promedios))
    lista_promedios.sort(reverse=True)
           
    lista_ids = []
    for promedio in lista_promedios:
        lista = [key for key, value in new_dic.items() if value['promedio'] == promedio]
        lista_ids = lista_ids + lista
        
    return lista_ids

ids_estrato1 = get_ordered_ids('1')
print(len(ids_estrato1))
ids_estrato2 = get_ordered_ids('2')
print(len(ids_estrato2))
ids_estrato3 = get_ordered_ids('3')
print(len(ids_estrato3))
ids_estrato4 = get_ordered_ids('4')
print(len(ids_estrato4))
ids_estrato5 = get_ordered_ids('5')
print(len(ids_estrato5))
```

```python
# código
def select_scholarships_estrato(porcentaje = 0.02,becas= 100):
    
    '''Genera un diccionario con la población escogida bajo el primer criterio propuesto.
    El primer input es el porcentaje de cada estrato por si se decide en algún momento cambiarlo.
    El segundo input de la función es el número de becas a asignar.'''
    
    lista_chosen_ids = []
    while True:
        for ids_estrato in [get_ordered_ids('1'),get_ordered_ids('2'),get_ordered_ids('3'),get_ordered_ids('4'),get_ordered_ids('5')]:
            datos= len(ids_estrato)
            posicion = int(porcentaje*datos)
            lista = ids_estrato[0:posicion]
            lista_filtered = [x for x in lista if x not in lista_chosen_ids]
            lista_chosen_ids = lista_chosen_ids + lista_filtered
            if len(lista_chosen_ids) >= becas:
                break
        if len(lista_chosen_ids) >= becas:
            break
        porcentaje = porcentaje + porcentaje
        
    lista_final = lista_chosen_ids[0:becas]

    diccionario_final = {key:value for key,value in diccionario_new.items() if key in lista_final}

    return diccionario_final

select_scholarships_estrato(becas = 6)
```

**Misión 5**

Desarrollar un algoritmo que permita conocer la población que sería becada según los criterios propuestos por el *segundo grupo* mencionado arriba.

```python
# código
def select_scholarships_region_genero(becas = 10):
    
    '''Genera un diccionario con la población escogida bajo el segundo criterio propuesto.
    El input de la función es el número de becas a asignar'''
    
    lista_regiones_unica = list(set(diccionario['region']))
    lista_generos_unica = list(set(diccionario['genero']))
    
    unicos = []
    for i in lista_regiones_unica:
        for j in lista_generos_unica:
            combinacion = (i,j)
            unicos.append(combinacion)
    
    becas_subgrupo = int(becas/len(unicos))
    
    lista_final = []
    
    for combinacion in unicos:
        
        region = combinacion[0]
        genero = combinacion[1]
        new_dic = {}
        
        for key, value in diccionario_new.items():
           if value['region'] == region and value['genero'] == genero:
               new_dic[key] = value
               
        lista_promedios = []
        for key, value in new_dic.items():
               lista_promedios.append(value['promedio'])
               
        lista_promedios = list(set(lista_promedios))
        lista_promedios.sort(reverse=True)
        
        lista_ids = []
        for promedio in lista_promedios:
            lista = [key for key, value in new_dic.items() if value['promedio'] == promedio]
            lista_ids = lista_ids + lista
            
        lista_ids_subgrupo = lista_ids[:becas_subgrupo]
        lista_final = lista_final + lista_ids_subgrupo
        
    diccionario_final = {key:value for key,value in diccionario_new.items() if key in lista_final}
    
    return diccionario_final

#Para este problema deben ser minimo 20 becas. Debido a que hay 20 subgrupos y tienen que ser igualmente distribuidos entre subgrupos.
select_scholarships_region_genero(becas = 20)
```

```python
#algoritmo que selecciona los aspirantes
def select_scholarships_edad_variable(becas = 10, min_edad = 18, max_edad = 30,variable='genero',diccionario_variable = diccionario_genero):
    
    '''
    becas = número_becas
    min_edad: edad mínima
    max_edad: edad máxima
    variable: es el criterio puede ser: "genero","estrato","region"
    diccionario_variable: dependiendo el criteroi se debe usar alguna de las plantillas dadas anteriormente'''
    
    #filtrar diccionario x edad
    new_dic_filtro_edad = {}
    
    for key, value in diccionario_new.items():
       if value['edad'] >= min_edad and value['edad'] <= max_edad:
           new_dic_filtro_edad[key] = value
    
    #aca saco cuentas variables le tocan de acuerdo al porcentaje dado
    dic_distribucion = {}
    for key,value in diccionario_variable.items():
        dic_distribucion[key] = int(value*becas)
       
    lista_ids_final = []
    
    for filtro,valor in dic_distribucion.items():
        
        new_dic = {}
        for key,value in new_dic_filtro_edad.items():
           if value[variable] == filtro:
               new_dic[key] = value
               
        #print(len(new_dic))
             
        lista_promedios = []
        for key, value in new_dic.items():
               lista_promedios.append(value['promedio'])
               
        lista_promedios = list(set(lista_promedios))
        lista_promedios.sort(reverse=True)
        
        lista_ids = []
        for promedio in lista_promedios:
            lista = [key for key, value in new_dic.items() if value['promedio'] == promedio]
            lista_ids = lista_ids + lista
            
        lista_ids = lista_ids[0:valor]

        lista_ids_final = lista_ids_final + lista_ids

    diccionario_final = {key:value for key,value in new_dic_filtro_edad.items() if key in lista_ids_final}
    
    return diccionario_final
```
